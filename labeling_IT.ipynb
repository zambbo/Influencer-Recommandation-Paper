```python
import pandas as pd

file_path = "../csvfile/결정트리/merge_data_label.csv"

IT_list = ['거리 측정기','높이 측정기','라즈베리 파이','마이크로 비트','산소포화도 측정기','소음 측정기','아두이노','온습도계','유수분 측정기','적외선 온도계','타이머']
NONIT_list = ['가방 만들기','가죽공예','무드등','미니어처','발난로','보석십자수','스크래치 북','양모펠트','오르골','전자저울','칼림바','캡슐 세제','펀치 니들','프랑스 자수']

```

```python
df = pd.read_csv(file_path,index_col=0)
df.head()
```

```python
def rowapply(row):
    ret_val = 0
    for nit in NONIT_list:
        if nit in row:
            ret_val = 0
    for it in IT_list:
        if it in row:
            ret_val = 1
    return ret_val
```

```python
df['IT/NONIT'] = df['Post'].apply(rowapply)
```
