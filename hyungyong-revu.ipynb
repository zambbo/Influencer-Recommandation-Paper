```python
import pandas as pd
file_path = './csvfile/non_it_dataset_after_processing.csv'

df = pd.read_csv(file_path,index_col=0)
df.head()
```

```python
import numpy as np
from konlpy.tag import Okt
```

```python
sympathy_median = df['Sympathy num'].median()
comment_median = df['Comment num'].median()
non_it_df = df[df['Sympathy num'] > sympathy_median]
non_it_df = non_it_df[non_it_df['Comment num'] > comment_median]
non_it_posts = non_it_df['Post']
```

```python
non_it_adjective_dict = dict()

for non_it_post in non_it_posts:
    pos = Okt().pos(non_it_post,norm=True,stem=True)

    okt_filtering = [x for x, y in pos if y in ['Adjective']]
    
    for adj in okt_filtering:
        if adj in non_it_adjective_dict:
            non_it_adjective_dict[adj] = non_it_adjective_dict[adj] + 1
        else:
            non_it_adjective_dict[adj] = 1
```

```python
sorted_non_it_adjective_dict = sorted(non_it_adjective_dict.items(),key=lambda x: x[1],reverse=True)
```
