```python
import pandas as pd
#file_path = './data_ITNONIT.csv'

df = pd.read_csv('./data_ITNONIT.csv',index_col=0)
df.head()
```

```python
import numpy as np
from konlpy.tag import Okt
```

```python
it_posts = df[df['IT/NONIT']==1]['Post']
nonit_posts = df[df['IT/NONIT']==0]['Post']
it_posts = it_posts.tolist()
nonit_posts = nonit_posts.tolist()
```

```python
it_adjective_dict = dict()

for it_post in it_posts:
    pos = Okt().pos(it_post,norm=True,stem=True)

    okt_filtering = [x for x, y in pos if y in ['Adjective']]
    
    for adj in okt_filtering:
        if adj in it_adjective_dict:
            it_adjective_dict[adj] = it_adjective_dict[adj] + 1
        else:
            it_adjective_dict[adj] = 1
```

```python
nonit_adjective_dict = dict()

for nonit_post in nonit_posts:
    pos = Okt().pos(nonit_post,norm=True,stem=True)

    okt_filtering = [x for x, y in pos if y in ['Adjective']]
    
    for adj in okt_filtering:
        if adj in nonit_adjective_dict:
            nonit_adjective_dict[adj] = nonit_adjective_dict[adj] + 1
        else:
            nonit_adjective_dict[adj] = 1
```

```python
sorted_it_adjective_dict = sorted(it_adjective_dict.items(),key=lambda x: x[1],reverse=True)
sorted_nonit_adjective_dict = sorted(nonit_adjective_dict.items(),key=lambda x:x[1],reverse=True)
```
