```python
import pandas as pd
file_path = './csvfile/mechalog.csv'

df = pd.read_csv(file_path,index_col=0)
df.head()
```

```python
import numpy as np
from konlpy.tag import Okt
```

```python
it_df = df[df['click'] >df['click'].median()]
it_posts = it_df['Post']
```

```python
it_adjective_dict = dict()

for it_post in it_posts:
    pos = Okt().pos(it_post,norm=True,stem=True)

    okt_filtering = [x for x, y in pos if y in ['Adjective']]
    
    for adj in okt_filtering:
        if adj in it_adjective_dict:
            it_adjective_dict[adj] = it_adjective_dict[adj] + 1
        else:
            it_adjective_dict[adj] = 1
```

```python
sorted_it_adjective_dict = sorted(it_adjective_dict.items(),key=lambda x: x[1],reverse=True)
```
